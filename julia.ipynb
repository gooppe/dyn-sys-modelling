```python
import numpy as np
import matplotlib.pyplot as plt

from itertools import product

plt.rcParams["figure.figsize"] = (20,20)
```

```python
def calc_julia(xs, ys, c, iters=256):
    r = (1 + np.sqrt(1 + 3 * np.abs(c))) / 2
    plot = np.ones((len(xs), len(ys))) * iters
    for i, x in enumerate(xs):
        for j, y in enumerate(ys):
            z = complex(x, y)
            for it in range(iters):
                z = z ** 2 + c
                if np.abs(z) > r:
                    plot[i, j] = it
                    break
    return plot / iters
```

```python
x = np.linspace(-1.5, 1.5, 1000)
y = np.linspace(-1.5, 1.5, 1000)
c = -0.74543 + 0.11301j

img = calc_julia(x, y, c)

plt.imshow(img)
plt.show()
```
